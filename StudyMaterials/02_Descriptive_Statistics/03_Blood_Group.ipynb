```python
# Frequency
blood_group_new_df = pd.DataFrame(blood_group_df.value_counts(), columns=['frequency'])
blood_group_new_df
```

```python
blood_group_new_df['proportion'] = blood_group_new_df['frequency'] / blood_group_new_df['frequency'].sum()
blood_group_new_df
```

```python
blood_group_new_df['percentage'] = (blood_group_new_df['frequency'] / blood_group_new_df['frequency'].sum()) * 100
blood_group_new_df
```

```python
blood_group_new_df.loc['O']
```

```python
blood_group_new_df.loc['Total'] = [blood_group_new_df['frequency'].sum(), blood_group_new_df['proportion'].sum(), blood_group_new_df['percentage'].sum()] 
blood_group_new_df
```
