```python
entries = len(df_red)
red = np.repeat('red', entries)
len(red)
```

```python
entries = len(df_white)
white = np.repeat('white', entries)
len(white)
```

```python
df_red['colour'] = red
df_red.head()
```

```python
df_white['colour'] = white
df_white.head()
```

```python
df_wine = df_red.append(df_white, ignore_index=True)
df_wine.tail()
```

```python
df_wine.to_csv('wine.csv', index=False)
```

```python
r=pd.plotting.scatter_matrix(df_wine, figsize=(15,15))
```

```python
df_wine.pH.describe()
```

```python
bin_edges = [2.72, 3.11, 3.21, 3.32, 4.01]
bin_names = ['high', 'moderately_high', 'medium', 'low']
```

```python
df_wine['acidity_levels'] = pd.cut(df_wine['pH'], bin_edges, labels=bin_names)
df_wine.head()
```

```python
df_wine.groupby('acidity_levels')['quality'].mean().plot(kind='line', grid=True)
```
